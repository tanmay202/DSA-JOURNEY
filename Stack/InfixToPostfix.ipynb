#Algorithm

1.Create an empty stack for operators
2.Scan the infix expression from left to right
3.For eaach symbol:
    a)Add each operand to the output
    b)If "(" comes then push it to stack
    c)")" pop from the stack to output until "(" is found
    d)Pop operator from stack to output while precedence of top >= current operator
    e)Push current operator to stack
4.After scanning pop all remaining operator from stack to output


```python
def precedence(op):
    if op == "+" or op == "-":
        return 1
    elif op == "*" or op == "/":
        return 2
    elif op == "^":
        return 3
    else:
        return 0

def inf_post(exp):
    stack = []
    output = []

    for ch in exp:
        if ch.isalnum():
            output.append(ch)

        elif ch == "(":
            stack.append(ch)

        elif ch == ")":
            while stack and stack[-1] != "(":
                output.append(stack.pop())
            stack.pop()

        else:
            while stack and precedence(stack[-1]) >= precedence(ch):
                output.append(stack.pop())
            stack.append(ch)

    # OUTSIDE loop
    while stack:
        output.append(stack.pop())

    return ''.join(output)
```
