Stack ADT:
#push(x)==> Inseert element x at the top pf the stack
#pop==>Remove and return the top element from the stack
#isEmpty()==>Retun true if stack is empty
#isFull==>Return true if Stack is full
#peek()/top()==>Return the top element without removing 
#size()==>Return the number of the elements in the stack

```python
#list based implementation

stack=[]
size=int(input("Enter the size of the stack:"))
for i in range(size):
    a=int(input(f"Enter {i+1} element:"))
    stack.append(a) #push
    print("AFter insertion stack:",stack)

# Peek (top element)
print("Top element:", stack[-1])

# Check if empty
print("Is empty:", len(stack) == 0)

#size
print("Size:",len(stack))

#pop
for i in range(size):
    stack.pop() #pop
    
    print("AFter pop stack:",stack)

# Check if empty
print("Is empty:", len(stack) == 0)

```

```python
#class Based implementation

class Stack:
    #constructor
    def __init__(self,max_size=100):
        self.stack=[]
        self.max_size=max_size
    
    #push
    def push(self,item):
        if self.size()>self.max_size:
            raise OverflowError("Stack overflow")
        self.stack.append(item)
    
    #peek
    def peek(self):
        if self.isEmpty():raise IndexError("Stack is empty")
        
        return self.stack[-1]
    
    #pop
    def Po(self):
        if self.isEmpty():
            raise IndexError("Stack is empty")
        return  self.stack.pop()


    #isEmpty
    def isEmpty(self):
        return self.size()==0
    
    #isFull
    def isFull(self):
        return len(self.stack) == self.max_size
    
    #size
    def size(self):
        return len(self.stack)
```

```python
s=Stack()
s.push(10)
s.push(20)
s.push(40)

for item in s.stack:
    print(item)

print(f"Top: {s.peek()}")
print(f"{s.isEmpty()}")
print(f"Size:{s.size()}")
```
