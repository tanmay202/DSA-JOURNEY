In this series, we will explore different patterns. I will try to solve each one and understand them completely without memorization—focusing on pure understanding.

```python
n=6
```

Pattern 1:
*****
*****
*****
*****
*****

```python
for i in range(n):
    for j  in range(n):
        print("*",end="")  #print add newline auto,end="" prevents the new line
    print()
        
```

Pattern 2 (increasing stars)
*
**
***
****
*****

```python
for i in range(1,n):
    for j in range(1,i+1):
        print("*",end="")
    print()
```

Pattern 3 (decreasing stars)
*****
****
***
**
*

```python
for i in range(1,n):
    for j in range(n,i,-1):
        print("*",end="")
    print()
```

Pattern 4 (Right-aligned triangle)
    *
   **
  ***
 ****
*****

```python
for i in range(1,n):
    #spaces
    for s in range(n-i):
        print(" ",end="")
    #star
    for j in range(1,i+1):
        print("*",end="")
    print()
```

Pattern 5 (Full Pyramid)
    *
   ***
  *****
 *******
*********

```python
for i in range(1,n):
    #space
    for s in range(n-i-1):
        print(" ",end="")
    #star
    for j in range(2*i-1):
        print("*",end="")
    print()
```

Pattern 6 (Inverted Pyramid)
*********
 *******
  *****
   ***
    *

```python
for i in range(1, n+1):
    # spaces
    for s in range(i - 1):
        print(" ", end="")
    
    # stars
    for j in range(1,2*(n - i)):
        print("*", end="")
    
    print()
```
